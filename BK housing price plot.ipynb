```python
import pandas as pd
import numpy as np
from matplotlib.pylab import plt
import geopandas as gpd
```

```python
zips = gpd.read_file('C:\\Users\\86746\\Desktop\\NYU\\PUI\\数据\\ZIP_CODE_040114.shp')
zips=zips.to_crs(epsg=4326)
```

```python
Queens=zips[zips['COUNTY']=='Kings']
```

```python
df=pd.read_csv('BK.csv')
df=df.loc[:,['avg price','geometry']]
```

```python
df
```

```python
df['geometry'] = gpd.GeoSeries.from_wkt(df['geometry'])
```

```python
df = gpd.GeoDataFrame(df,geometry=df['geometry'])
```

```python
fig,ax = plt.subplots(nrows=1, ncols=1,figsize=(20,10))
Queens['geometry'].plot(ax=ax,color='white', edgecolor='black')
df.plot(ax=ax, column='avg price', k=10, scheme='Quantiles',  legend=True)
plt.show()
```
